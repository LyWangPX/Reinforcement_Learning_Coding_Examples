```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rewards = 0
dt = 1
x = 0.15
sigma = 0.2
theta = 0.15
data = [x]
for step in range(200):
    dW = dt*dt*np.random.normal()
    dx = -theta * x * dt + sigma * dW
    x += dx
    data.append(x)
plt.plot(data)
```

```python
class Ornstein_Uhlenbeck_Process:
    def __init__(self, dt = 0.1):
        self.theta = 0.15
        self.sigma = 0.2
        self.dt = dt
        self.x = 0

    def step(self):
        dW = self.dt**2*np.random.normal()
        dx = -self.theta * self.x * self.dt + self.sigma*dW
        self.x += dx
        return self.x
```

```python
random_process = Ornstein_Uhlenbeck_Process()
data = []
for _ in range(200):
    data.append(random_process.step())
plt.plot(data)
```
